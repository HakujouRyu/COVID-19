# Setting up Different DFs

- usc_time = A set of US confirmed CASES with a datetime index
- usd_tome = A set of US confirmed DEATHS with a datetime index
- usr_time = A set of US confirmed RECOVERIES with a datetime index


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
confirmed = pd.read_csv( "../csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv")
deaths = pd.read_csv("../csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv")
recovered = pd.read_csv("../csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv")
cases_g = pd.read_csv("../csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
deaths_g = pd.read_csv("../csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
```

# Confirmed Cases

```python
confirmed.head()
```

### USA

```python
usCases = confirmed[confirmed["Country/Region"] == "US"]
```

```python
usCases.head()
```

```python
usc_time = usCases.drop(columns=["Country/Region", "Lat","Long"]).T
```

```python
usc_time.columns = usc_time.iloc[0]
```

```python
usc_time = usc_time.drop('Province/State')
```

```python
usc_time.index = pd.to_datetime(usc_time.index)
```

```python
usc_time.isna().sum().sort_values(ascending=False)
```

```python
usc_time.head()
```
